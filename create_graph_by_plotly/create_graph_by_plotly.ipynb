```python
tokyo_meteorological_data_2022 = pd.read_csv('/content/2022年東京都の気象データ.csv', encoding='shift_jis')
```

```python
# 2つ目の軸を表示するか否かを設定
fig = make_subplots(specs=[[{"secondary_y": True}]])

# 第一軸のグラフ
fig.add_trace(go.Line(
    x=tokyo_meteorological_data_2022["年月日"],
    y=tokyo_meteorological_data_2022["平均気温(℃)"],
    name ="2022年の東京の平均気温"
       ))

# 第二軸のグラフ
fig.add_trace(go.Line(
    x=tokyo_meteorological_data_2022["年月日"],
    y=tokyo_meteorological_data_2022["日照時間(時間)"],
    name ="2022年の東京の日照時間(時間)"),
       secondary_y=True)

# グラフのタイトルと各軸のタイトル設定
fig.update_layout(
    # スマホのような狭い画面で表示する場合に凡例の位置を見やすい位置に変える
    # legend=dict(
    #     x=0,
    #     y=1.1,
    #     xanchor='left',
    #     yanchor='top',
    #     orientation='h',
    # ),
    title="<b>2022年の東京の平均気温と日照時間(時間)の関係</b>"
)
fig.update_yaxes(title_text="<b>平均気温(℃)</b>", secondary_y=False)
fig.update_yaxes(title_text="<b>日照時間(時間)</b>", secondary_y=True)
fig.update_xaxes(title="<b>年月</b>")

# このまま出力すると文字化けしてしまうので日本語のフォントを設定する
font = font_manager.FontProperties(
    family="IPAGothic",
    size=30
)
fig.show()
```
