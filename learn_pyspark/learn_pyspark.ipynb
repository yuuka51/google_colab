```python
from pyspark.sql.functions import current_timestamp

# 既存のDataFrameを作成する
df = spark.createDataFrame([(1, "foo"), (2, "bar"), (3, "baz")], ["id", "value"])

# 現在時刻を持つ列を作成する
current_time_col = current_timestamp().alias("current_time")
print(current_time_col)

# 既存のDataFrameに現在時刻を持つ列を追加する
df_with_time = df.withColumn("current_time", current_time_col)

# 結果を表示する
df_with_time.show()
```

<h3>DataFrame内のデータを全て計算する方法</h3>
<br>
パターン①<br>
同一の行の指定したカラムで計算されるため、一行で全てのデータの計算ができる

```python
diff = df['df1'] - df['df2']
diff
```

パターン②<br>
applyを使って1行ごとに計算させたい場合に関数を作成することで計算できる

```python
def setDiff(row):
    row['diff'] = row['df1'] - row['df2']
    return row
```

```python
df_diff = df.apply(setDiff, axis=1)
df_diff
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Sparkセッションを作成
spark = SparkSession.builder.getOrCreate()

# サンプルのDataFrameを作成
data = [("Alice", [10, 20, 30]), ("Bob", [40, 50, 60]), ("Charlie", [70, 80, 90])]
df = spark.createDataFrame(data, ["Name", "Numbers"])

# リストの0番目を取得する
df_with_first_element = df.withColumn("FirstNumber", col("Numbers").getItem(0))
# ↑と↓で同じ結果。↓の方がPythonのよく書く書き方。↑はJavaライクっぽい
df_with_first_element = df.withColumn("FirstNumber", col("Numbers")[0])


# 結果を表示
df_with_first_element.show()
```
